```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Proxy1\Downloads\train.csv")
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',hue='Pclass',errorbar=None,palette='coolwarm')
plt.title("Survival Rate by Passenger Class")
plt.xlabel('Passenger Class')
plt.ylabel("Survival Rate")
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Pclass',palette='dark',alpha=0.6)
plt.title("age and fare relation")
plt.xlabel('AGE')
plt.ylabel("Fare")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare','Pclass','Survived']],hue='Survived',palette='husl')

plt.show()
```

```python
df_numeric = df.select_dtypes(include=['number'])
correlation_matrix = df_numeric.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
