```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tips=sns.load_dataset('tips') #Load the 'tips' dataset
tips.head()
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="lavender")
plt.title("Distribution of Total Bills")
plt.show()
```

```python
#Kernel density Estimation (KDE) visulaizes the propability density
sns.kdeplot(data=tips,x="total_bill",fill=True,color="purple")
plt.title("KDE plot of Total bills")
plt.show()
```

```python
#Bar plot
sns.barplot(data=tips,x='day',y='total_bill',palette='viridis',legend=False)
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box Plot of Total Bills by Day and Gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter Plot of Total Bill vs. Trip")
plt.show()
```

```python
# PAIR PLOT
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.show()
```

```python
#line plot
sns.lineplot(data=tips,x='size',y='tip',marker='o',color='green')
plt.title("Line Plot of Party Size vs. Tip")
plt.show()
```

```python
#HEAT MAP
numeric_cols=tips.select_dtypes(include=['number']).columns
cor_matrix=tips[numeric_cols].corr()
sns.heatmap(cor_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
