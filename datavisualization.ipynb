```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,color='blue',linestyle='dashed',linewidth=2,marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[5000,10000,18000,25000,30000,35000]
plt.plot(x,y)
plt.title("savings account ")
plt.plot(x,y,color='green',linestyle='dashed',linewidth=2,marker='o')
plt.xlabel("Months")
plt.ylabel("Svaings amount")
plt.show()
```

```python
category=['A','B','C','D','E',]
values=[10,15,7,12,9]
plt.bar(category,values,color='skyblue')
plt.title("Basic bar chart")
plt.xlabel("category")
plt.ylabel("values")
plt.show()
```

```python
plt.bar(category,values,color=['red','pink','green','yellow','purple'],edgecolor='black')
plt.show()
```

```python
plt.bar(category,values,width=0.5)
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,35,37,40,45,47,48,51,55,60,65]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age distribution of customers")
plt.xlabel("Age ranges")
plt.ylabel("Frequency")
plt.grid(True)


```

```python
import matplotlib.pyplot as plt

x1 = ['python', 'java', 'c++', 'ruby']
x2 = [40.0, 30.0, 20.0, 10.0]

plt.pie(x2, labels=x1, colors=['yellow', 'skyblue', 'green', 'pink'], autopct='%1.1f%%',startangle=90)
plt.legend()
plt.title('Programming Language usage')
plt.show()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter plot example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
